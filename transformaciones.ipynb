# **Grafico de una traslación**



```python
import matplotlib.pyplot as plt

coordenada_inicial = input("Ingresar la coordenada inicial x,y:")
x1,y1 = coordenada_inicial.split(",")
x1 = float(x1)
y1 = float(y1)

coordenada_final = input("Ingresar la coordenada final x,y:")
x2,y2 = coordenada_final.split(",")
x2 = float(x2)
y2 = float(y2)

traslacion = input("Ingresar la coordenada x,y:")
tx,ty = traslacion.split(",")
tx = float(tx)
ty = float(ty)

eje_x = [x1,x2]
eje_y = [y1,y2]

eje_tx = [x1+tx, x2+tx]
eje_ty = [y1+ty, y2+ty]

plt.plot(eje_x, eje_y)
plt.plot(eje_tx, eje_ty)
plt.title("Traslación 2D")
plt.legend(("Inicial", "Trasladado"))
plt.show()
```

# **Gráfico de una rotación**

```python
import matplotlib.pyplot as plt
import math

coordenada_final = input("Ingrese la coordenada de inicio:")
x,y = coordenada_final.split(",")
x=float(x)
y=float(y)
angulo = float(input("Ingrese el angulo en sexagesimales: "))
angulo = (angulo*math.pi/180)

eje_x = [0,x]
eje_y = [0,y]

eje_rx = [0,x*math.cos(angulo)-y*math.sin(angulo)]
eje_ry = [0,x*math.sin(angulo)+y*math.cos(angulo)]

plt.plot(eje_x, eje_y)
plt.plot(eje_rx,eje_ry)
plt.title("Rotación en 2D")
plt.legend(("Inicial","Rotación"))

plt.show()
```

# **Gráfico de una escalación**

```python
import matplotlib.pyplot as plt

coordenada_inicial = input("Ingresar la coordenada inicial x,y:")
x1,y1 = coordenada_inicial.split(",")
x1 = float(x1)
y1 = float(y1)

coordenada_final = input("Ingresar la coordenada final x,y:")
x2,y2 = coordenada_final.split(",")
x2 = float(x2)
y2 = float(y2)

s1 = float(input("Ingresar la escala:"))

eje_x = [x1,x2]
eje_y = [y1,y2]

eje_dx = x2-x1
eje_dy = y2-y1

if s1 >= 1: 
  eje_sx = [x1, (x2)+(eje_dx*(s1-1))]
  eje_sy = [y1, (y2)+(eje_dy*(s1-1))]
  plt.plot(eje_sx, eje_sy)
  plt.plot(eje_x, eje_y)
  plt.legend(("Escalación", "Inicial"))
else:
  eje_sx = [x1, (x2)-(eje_dx)*s1]
  eje_sy = [y1, (y2)-(eje_dy)*s1]
  plt.plot(eje_x, eje_y)
  plt.plot(eje_sx, eje_sy)
  plt.legend(("Inicial", "Escalación"))

plt.title("Escalación 2D")
plt.show()
```

# **Gráfico en 3D**

```python
3import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
z = np.linspace(0, 15, 100)
x = np.sin(z)
y = np.cos(z)
traslacionX = float(input("Ingrese la traslación en el eje X: "))
traslacionY = float(input("Ingrese la traslación en el eje Y: "))
traslacionZ = float(input("Ingrese la traslación en el eje Z: "))
x_trasladado = []
y_trasladado = []
z_trasladado = []
for i in x:
    x_trasladado.append(i+traslacionX)
for i in y:
    y_trasladado.append(i+traslacionY)
for i in z:
    z_trasladado.append(i+traslacionZ)
ax.plot3D(x, y, z, 'red')
ax.plot3D(x_trasladado, y_trasladado, z_trasladado, 'green')
ax.view_init(-140, 60)
plt.title("Grafica 3D")
plt.legend(("Inicial", "Trasladado"))
plt.show()
```
